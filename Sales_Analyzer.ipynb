## Reading the file


```python
file = open("sales_data_local.csv", "r")
lines = file.readlines()
print(lines)
```

## New data


```python
newData = []
for line in lines[1: ]:
    field = line.strip().split(",")
    newData = []
    for line in lines[1:]:
        field = line.strip().split("\t")
        field[5] = int(field[5])
        field[6] = int(field[6])
        newData.append(field)

    print(newData)
    field[6] = int(field[6])
    newData.append(field)

print(newData)
```

```python
#Query 1

total_revenue = 0

for row in newData:
    quantity = row[5]
    price = row[6]
    total_revenue += quantity * price

print("Total Revenue: ₹", total_revenue)
```

```python
# Query 2

total_orders = len(newData)
print("Total Numbers of Orders ", total_orders)
```

```python
# Query 3

total_quantity = 0

for row in newData:
    total_quantity += row[5]

print("Total Quanitity Sold: ", total_quantity)
```

```python
# Query 4

catergory_revenue = {}

for row in newData:
    catergory = row[4]
    revenue = row[5]*row[6]

    if catergory in catergory_revenue:
        catergory_revenue[catergory] += revenue
    else:
        catergory_revenue[catergory] = revenue
    
print("Revenue by Category", catergory_revenue)
for cat, rev in catergory_revenue.items():
    print(f" {cat}: ₹{rev}")

```

```python
# Query 5

product_revenue = {}

for row in newData:
    product = row[3]
    revenue = row[5]*row[6]

    if product in product_revenue:
        product_revenue[product] += revenue
    else:
        product_revenue[product] = revenue
    
print("Revenue by Products", product_revenue)
for pro, rev in product_revenue.items():
    print(f" {cat}: ₹{rev}")
```

```python
# Query 6

product_quantity = {}

for row in newData:
    product = row[3]
    quantity = row[5]

    if product in product_quantity:
        product_quantity[product] += quantity
    else:
        product_quantity[product] = quantity

highest_Sold = None
max_quanity = 0

for product, qty in product_quantity.items():
    if qty > max_quanity:
        max_quanity = qty
        highest_Sold = product

print("Highest Sold Product: ", highest_Sold,"\nQuantity Sold: ", max_quanity)
```

```python
#Query 7

customer_revenue = {}

for row in newData:
    customer = row[2]
    revenue = row[5]*row[6]

    if customer in customer_revenue:
        customer_revenue[customer] += revenue
    else:
        customer_revenue[customer] = revenue

top_customer = None
max_revenue = 0

for cust, rev in customer_revenue.items():
    if rev>max_revenue:
        max_revenue = rev
        top_customer = cust
print("top Customer: ", top_customer, "\n Max Revenue: ₹", max_revenue)
```

```python
#Query 8

daily_revenue = {}

for row in newData:
    date = row[1]
    revenue = row[5]*row[6]

    if date in daily_revenue:
        daily_revenue[date] += revenue
    else:
        daily_revenue[date] = revenue

print("daily revenue")
for date, rev in daily_revenue.items():
    print(f" {date}: ₹{rev}")
```

```python
# Query 9

from datetime import datetime

start = input("Enter Start Date (YYYY-MM-DD)") 
end = input("Enter End Date (YYYY-MM-DD)") 

start_date = datetime.strptime(start, "%Y-%m-%d")
end_date = datetime.strptime(end, "%Y-%m-%d")

total_dateRev = 0

for row in newData:
    order_date = datetime.strptime(row[1], "%d-%m-%Y")

    if start_date <= order_date <= end_date:
        total_dateRev += row[5]*row[6]

print("Revenue from", start, "and", end,"is ₹", total_dateRev) 
```

```python
# Query 10
elecFile = open("electricals.csv", "w")

elecFile.write("OrderID,Date,Customer,Product,Category,Quantity,Price\n")

for row in newData:
    if row[4] == "Electronics":
        row_str = ",".join([str(item) for item in row])
        elecFile.write(row_str + "\n")
elecFile.close()

print("all electritical items saved to electricals.csv")
```
